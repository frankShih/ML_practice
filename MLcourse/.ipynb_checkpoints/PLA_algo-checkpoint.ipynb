```python
import numpy as np


def naive_pla(datas):  
    w = datas[0][0]
    iteration = 0
    while True:
        iteration += 1
        false_data = 0

        for data in datas:
            t = np.dot(w, data[0])
            if np.sign(data[1]) != np.sign(t):
                error = data[1]  
                false_data += 1
                w += error * data[0]
        print 'iter%d (%d / %d)' % (iteration, false_data, len(datas))
        if not false_data:
            break
    return w
```

```python
import matplotlib.pyplot as plt  
import numpy as np

ITERATION = 100;
WEIGHT = [100, 2, -1];


def createData():
    return np.array([
            [1.0, 1.0, 2.0, 1],
            [1.0, 2.0, 3.0, 1],
            [1.0, 3.0, 4.0, 1],
            [1.0, 3.0, 2.0, 1],
            [1.0, 5.0, 3.0, 1],
            [1.0, 2.0, 4.0, 1],
            [1.0, 6.0, 7.0, -1],
            [1.0, 7.0, 8.0, -1],
            [1.0, 9.0, 10.0, -1],
            [1.0, 6.0, 9.0, -1],
            [1.0, 7.0, 5.0, -1],
            [1.0, 9.0, 7.0, -1],
#            [threshold, x, y, label]
           ])    


def sign(W, dataList):      #
    product = 0;
    for i in range(len(W)):
        product += W[i] * dataList[i]
    if product > 0: return 1
    else: return -1

def renewW(W, trainData):   #W
    signResult = sign(W, trainData)
    if signResult == trainData[-1]: return W
    for k in range(len(W)):
        W[k] = W[k] + trainData[-1]*trainData[k]
    return W

def trainW(W, trainDatas):  #W
    newW = []
    for num in range(ITERATION):
        index = num % len(trainDatas)
        newW = renewW(W, trainDatas[index])
        print(newW)
        if num%10 == 0:
            plotBoundary(trainDatas, 6)
    return newW

def predictTestData(W, trainDatas, testDatas):  #
    W = trainW(W, trainDatas)
    print(W)
    for i in range(len(testDatas)):
        result = sign(W, testDatas[i])
        print(result)

def plotBoundary(dataset, index=0):
    fig = plt.figure()
    ax = fig.gca()
    plt.xlabel("X")
    plt.ylabel("Y")    
    ax.set_xticks(np.arange(-10, 10, 1))
    ax.set_yticks(np.arange(-10, 10, 1))
    ax.set_xlim([-10, 10])
    ax.set_ylim([-10, 10])
    plt.gca().set_aspect('equal', adjustable='box')
    plt.rcParams['figure.figsize'] = [7, 7]
    plt.grid()
    
    
    plt.scatter(dataset[:index,1], dataset[:index, 2], c="g", alpha=0.5, 
                marker= 'o', label="Positive")
    plt.scatter(dataset[index:,1], dataset[index:, 2], c="r", alpha=0.5, 
                marker= 'x', label="Negative")
    plt.legend(loc=2)
    plt.plot([0, WEIGHT[1]], [0, WEIGHT[2]], 'bo-')
    
    xx = np.arange(-10, 10, 0.5)
#     W[0]+W[1]*xx+W[2]*yy = 0
#     W[1]*xx+W[2]*yy = -1*W[0]
#     W[2]*yy = -1*W[0] - W[1]*xx
#     yy = (-1*W[0] - W[1]*xx)/W[2]
    yy = (-WEIGHT[1]*xx - WEIGHT[0])/WEIGHT[2];
    plt.plot(xx, yy)
  

    plt.show()

    
dataSet = createData();        
trainW(WEIGHT, dataSet)
# plotBoundary(dataSet, 6)
```
